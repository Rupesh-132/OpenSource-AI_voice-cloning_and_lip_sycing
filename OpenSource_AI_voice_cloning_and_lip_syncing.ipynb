```python
import subprocess
import whisper
class AudioProcessor:
    def __init__(self):
      self.video_path = None
      self.whisper_text = None
      self.whisper_language = None

    def extract_audio(self, video_path):
      self.video_path = video_path
      if self.video_path:
        ffmpeg_command = f"ffmpeg -i '{self.video_path}' -acodec pcm_s24le -ar 48000 -q:a 0 -map a -y 'main_audio_output.wav'"
        subprocess.run(ffmpeg_command, shell=True)
      else:
        print("No video uploaded. Please upload a video first.")

    def convert_to_whisper(self):
      if self.video_path:
          model = whisper.load_model("base")
          result = model.transcribe("/content/main_audio_output.wav")
          self.whisper_text = result["text"]
          self.whisper_language = result['language']
          print("Whisper text:", self.whisper_text)
      else:
        print("No video uploaded. Please upload a video first.")


# Example usage
audio_processor = AudioProcessor()

# Assuming 'video_path' variable exists (you can set it to the actual video path)
video_path = 'path_to_your_video.mp4'  # Replace 'path_to_your_video.mp4' with the actual video path
audio_processor.extract_audio(video_path)
audio_processor.convert_to_whisper()
```

```python
from openai import OpenAI
whisper_text = "Everybody has a turn back moment. You have a moment where you could go forward or you can give up. But the thing you have to keep in mind before you give up is that if you give up the guarantee is it will never happen. That's the guarantee of quitting that it will never happen no way under the sun. The only way the possibility remains that it can happen is if you never give up no matter what."
client = OpenAI(api_key = "")
completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": f"translate the texts to Hindi {whisper_text}"}
  ]
)
translated_text = completion.choices[0].message
print(translated_text)
```

```python
translated_text = translated_text.content
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
      # Set a really small chunk size, just to show.
          chunk_size=250,
              chunk_overlap=0,
                  length_function=len,
                      is_separator_regex=False,
                      )
```

```python
texts = text_splitter.create_documents([translated_text])

final_chunks = []
for i in range(0,len(texts)):
  final_chunks.append(texts[i].page_content)
```

```python
final_chunks
```

```python

import torch
# @title Voice synthesis for text chunks
def audio_synthesis(text, file_name):
  tts.tts_to_file(
      text,
      speaker_wav='/content/main_audio_output.wav',
      file_path=file_name,
      language="hi"
  )
  return file_name
file_names = []
for i in range(len(final_chunks)):
    file_name = audio_synthesis(final_chunks[i], f"output_synth_audio_{i}.wav")
    file_names.append(file_name)
```

```python
from pydub import AudioSegment

def combine_audio_files(audio_file1, audio_file2, output_path):
    # Load audio files
        audio1 = AudioSegment.from_file(audio_file1)
        audio2 = AudioSegment.from_file(audio_file2)

        # Combine the two audio files
        combined_audio = audio1 + audio2

        # Export the combined audio to a new file
        combined_audio.export(output_path, format="mp3")


# Replace these paths with the paths to your audio files
audio_file1 = "/content/output_synth_audio_0.wav"
audio_file2 = "/content/output_synth_audio_1.wav"

# Specify the output path for the combined audio file
output_file = "combined_audio.mp3"

combine_audio_files(audio_file1, audio_file2, output_file)
print(f"Audio files combined and saved to {output_file}")
```

```python
import os

class VideoGenerator:
    def __init__(self):
      self.video_path = None
      self.pad_top = 0
      self.pad_bottom = 15
      self.pad_left = 0
      self.pad_right = 0
      self.rescale_factor = 1

    def set_video_path(self, video_path):
      self.video_path = video_path

    def set_padding(self, top, bottom, left, right):
      self.pad_top = top
      self.pad_bottom = bottom
      self.pad_left = left
      self.pad_right = right

    def set_rescale_factor(self, factor):
      self.rescale_factor = factor

    def generate_video(self, audio_path, output_path):
      if self.video_path:
        video_path_fix = f"'../{self.video_path}'"
        command = f"python inference.py --checkpoint_path 'checkpoints/wav2lip_gan.pth' --face {video_path_fix} --audio {audio_path} --pads {self.pad_top} {self.pad_bottom} {self.pad_left} {self.pad_right} --resize_factor {self.rescale_factor} --nosmooth --outfile {output_path}"
        os.system(command)
      else:
         print("No video path provided. Please set the video path before generating the video.")

video_generator = VideoGenerator()
video_generator.set_video_path("Steve Harvey- Inspirational Speech _ Motivational Short Video _ Incredible You.mp4")
video_generator.generate_video("/content/combined_audio.mp3", "/content/modified_video_output.mp4")
```
